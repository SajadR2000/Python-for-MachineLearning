## Loading dataset
Load the IMDB-Movie-Data.csv file into a pandas dataframe and show its first 5 rows. Each row of this dataframe corresponds to one film.

```python
df = pd.read_csv('IMDB-Movie-Data.csv')
df.head(5)
```

## Calculating Profit
Add a column to the dataframe that calculate the profit of the movies. (Suppose the inflation rate is 3% and 1990 as the base year)

```python
df['Profit'] = df['Revenue (Millions)'] / (1.03 ** (df['Year'] - 1990))
df.head()
```

## Nolan Movies
Filter Christopher Nolan movies and sort them in a descending order with respect to their profits then show the result.

```python
nolan = df[df['Director'] == 'Christopher Nolan']
nolan = nolan.sort_values(by=['Profit'], ascending=False)
nolan
```

## Favorite Genres
Calculate the revenues of different genres in a dataframe and sort them in a descending order. Then show the result.

```python
ndf = pd.DataFrame([], columns=df.columns)
for i in range(df.shape[0]):
    temp = df.iloc[i]['Genre'].split(',')
    for j in range(len(temp)):
        t = df.iloc[i].copy()
        t['Genre'] = temp[j]
        ndf = ndf.append(t)

x = ndf.groupby(['Genre']).sum()
x.sort_values(by=['Profit'], ascending=False,inplace=True)
x
# The following code is the more efficient way but doesn't work on my version of pandas!
# a = df.copy()
# a['Genre'] = a['Genre'].apply(lambda x:x.split(','))
# a.explode('Genre').sort_values(by=['Profit'], ascending=False,inplace=True)
```

## Actors/Actresses Activity
For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
actors = df['Actors'].copy()
actors = actors.to_frame()
# Add ',' at the start of every actors list in order to prevent joining last actor of i-th movie with first actor of (i+1)-th movie.
actors['Actors'] = ',' + actors['Actors']
actorsString = actors['Actors'].sum()
actorsString = actorsString[1:]
actorsList = actorsString.split(',')
for i in range(len(actorsList)):
    if actorsList[i][0] == ' ':
        actorsList[i] = actorsList[i][1:]
actors_df = pd.DataFrame(actorsList, columns=['Actors'])
result = actors_df['Actors'].value_counts().rename_axis('Actors').reset_index(name='#ofMovies')
print(result.columns)
result.sort_values(by=['#ofMovies', 'Actors'], ascending=[False, True], inplace=True)
result.reset_index()
```
